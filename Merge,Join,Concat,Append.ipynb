```python
import pandas as pd
```

```python
df1=pd.DataFrame([[1,'hamza',3.5],[2,'farhan',5.0],[3,'vicky',4.8]],columns=['id','name','rating'])
```

```python
df2=pd.DataFrame([[1,'kwl',45],[2,'nankna',50],[3,'kwl',38],[3,'dfs',45]],columns=['id','adress','score'])
```

```python
#if two columns same good else use left_on or right_on or left_index or right_index
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
#if two columsn same use lsufix to change one columns name
df2.join(df1,sort=True,lsuffix='a')
```

```python
#they add data in verticaly or horizently as we want
pd.concat([df1,df2],axis=1)
```

```python
#add data in vertically
df1.append(df2,ignore_index=False,sort=True)
```

```python
#include every thing which mathced in both same columns in this case id
df1.join(df2,how='inner',lsuffix='a')
```

```python
#include everything of dataframes if is or not
df1.join(df2,how='outer',lsuffix='a')
```

```python
#include everything of left columns
df1.join(df2,how='left',lsuffix='a')
```

```python
#include everything of right column
df1.join(df2,how='right',lsuffix='a')
```
